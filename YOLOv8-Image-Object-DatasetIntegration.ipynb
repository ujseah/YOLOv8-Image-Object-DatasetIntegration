```python
import os

datasets = [
    "/earphones(1)",
    "/earphones(2)",
    "/keys(1)",
    "/keys(2)",
    "/phones(1)",
    "/rings(1)",
    "/rings(2)",
    "/rings(3)",
    "/wallet(1)",
    "/wallet(2)",
    "/wallet(3)",
    "/wallet(4)",
    "/watches(1)",
    "/watches(2)",
    "/watches(3)"
]

class_IDs = ("earphones", "keys", "phones", "rings", "wallets", "watches")

class_mapping = {}
for dataset_index, dataset in enumerate(datasets):
    for class_index, class_id in enumerate(class_IDs):
        if class_id[:-1] in dataset:
            class_mapping[dataset_index] = class_index
            break

print("Generated class_mapping:", class_mapping)

for dataset_index, dataset in enumerate(datasets):
    dataset_dir = os.path.join('/content', dataset.strip('/'))

    new_class_id = class_mapping[dataset_index]

    for split in ['train', 'valid', 'test']:
        labels_dir = os.path.join(dataset_dir, split, 'labels')

        if not os.path.exists(labels_dir):
            print(f"Directory not found: {labels_dir}")
            continue

        for label_file in os.listdir(labels_dir):
            file_path = os.path.join(labels_dir, label_file)

            with open(file_path, 'r') as f:
                lines = f.readlines()

            updated_lines = []
            for line in lines:
                parts = line.strip().split()
                if len(parts) == 0:
                    continue

                old_class_id = int(parts[0])
                parts[0] = str(new_class_id)
                updated_lines.append(' '.join(parts) + '\n')

            with open(file_path, 'w') as f:
                f.writelines(updated_lines)
```

```python
import yaml

# Define the dataset configuration
data = {
    'train': '/content/final_dataset/train/images',
    'val': '/content/final_dataset/valid/images',
    'test': '/content/final_dataset/test/images',  # Optional
    'nc': 6,  # Number of classes
    'names': ["earphones", "keys", "phones", "rings", "wallets", "watches"]  # Class names
}

# Save to a data.yaml file
with open('/content/final_dataset/data.yaml', 'w') as f:
    yaml.dump(data, f)

print("data.yaml file created!")
```

```python
from ultralytics import YOLO

# Load YOLOv8 model
model = YOLO('yolov8n.pt')

# Train the model
model.train(data='/content/final_dataset/data.yaml', epochs=10, imgsz=640, batch=16)
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
from pathlib import Path

# Load the YOLOv8 model
model = YOLO("/content/runs/detect/train/weights/best.pt")  # Path to your best.pt file

# Test image path
image_path = "/content/runs/detect/keystest2.jpg"  # Path to your test image

# Run inference
results = model.predict(source=image_path, conf=0.3, save=True)

# Automatically detect the output folder
output_folder = Path(results[0].path).parent  # Finds the directory where YOLOv8 saved the results

# Path to the processed image with bounding boxes
output_image_path = output_folder / Path(image_path).name

# Load the image with bounding boxes
output_image = cv2.imread(str(output_image_path))
output_image = cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB)

# Display the image
plt.figure(figsize=(10, 10))
plt.imshow(output_image)
plt.axis("off")
plt.show()
```
